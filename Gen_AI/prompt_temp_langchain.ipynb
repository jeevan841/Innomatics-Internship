```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

```python
#part-01:
template = "Explain {topic} in simple terms so beginners can understand."

prompt_temp = PromptTemplate(
    input_variables=["topic"],
    template=template,
)

# Testing
print(prompt_temp.format(topic="Artificial Intelligence"))
print(prompt_temp.format(topic="Machine Learning"))
print(prompt_temp.format(topic="Deep Learning"))
```

```python
#part-02:
multi_temp = "Explain {topic} to {audience} with a {tone} tone."

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_temp
)

# Testing
print(multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
#part-03:
# Teaching
teaching_template = PromptTemplate(
    input_vari=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview
interview_template = PromptTemplate(
    input_vari=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling
story_template = PromptTemplate(
    input_vari=["topic"],
    template="Explain {topic} as a story"
)

# Runnning
topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python
#part-04:
def get_prompt(role, topic):

    system_roles = {
        "teacher": "You are a helpful teacher.",
        "interviewer": "You are a strict interviewer.",
        "motivator": "You are an inspiring mentor."
    }

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_roles[role]),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)

#  Testing
messages = get_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

```python
#part-05:
def validate_inputs(audience, tone):

    given_audiance = ["beginner", "intermediate", "expert"]
    given_tone = ["formal", "casual", "fun"]

    if audience not in given_audiance:
        print("wrong audience! Setting to 'beginner'")
        audience = "beginner"

    if tone not in given_tone:
        print("wrong tone! Setting to 'formal' tone")
        tone = "formal"

    return audience, tone
```

```python
#part-06:
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=styles[style]
    )

    return template.format(topic=topic, audience=audience, tone=tone)

# Testing
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
#part-07:
reuse_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("ML", "expert", "casual"),
    ("Data Science", "beginner", "formal"),
    ("NLP", "expert", "fun")
]

for topic, audience, tone in inputs:
    print(reuse_template.format(topic=topic, audience=audience, tone=tone))
```
