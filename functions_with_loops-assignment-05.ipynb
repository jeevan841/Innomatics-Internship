```python
#Social Media – Post Engagement Analyzer
```

```python
def likes_analyse(l_list):
    total_likes = 0
    
    for likes in l_list:
        total_likes += likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes = [200, 300, 150, 400]
likes_analyse(likes)
```

```python
#Agriculture – Rainfall Adequacy Checker
```

```python
def rainfall_check(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain
    average = total / len(rainfall_list)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_data = [70, 75, 68, 80, 67]
required = 70
rainfall_check(rainfall_data, required)
```

```python
#Social Media – Duplicate Account Detection
```

```python
def duplicate_accounts(usernames):
    
    unique_usernames = set(usernames)
    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

accountlist = ["user1", "user2", "user3", "user2", "user4"]
duplicate_accounts(accountlist)
```

```python
#Agriculture – Premium Crop Price Filter
```

```python
def crop_filter(prices):
    
    premium_crops = []
    
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

crop_prices = [1500, 2500, 1800, 3200, 900]
crop_filter(crop_prices)
```

```python
#E-Learning – Student Attendance Eligibility System
```

```python
def attendance_check(attendance_list):
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    for status in attendance_list:
        if status == 1:  
            attended_classes += 1
    
    percentage = (attended_classes / total_classes) * 100
    
    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

attendance = [1, 1, 1, 0, 1]  
attendance_check(attendance)
```

```python
#Smart Electricity Bill Analyzer
```

```python
def bill_analysis(units):
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status

units = 250
bill, status = bill_analysis(units)

print("Units Consumed:", units)
print("Total Bill: ₹", bill)
print("Usage Classification:", status)
```
