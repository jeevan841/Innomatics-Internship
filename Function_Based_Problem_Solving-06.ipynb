```python
#Smart Parking Lot Management System
```

```python
def management_parking(capacity, parks):
    current_parked = 0
    peak_parking = 0
    for park in parks:
        
        if park == "IN":
            current_parked += 1
        elif park == "OUT":
            current_parked -= 1
        if current_parked > peak_parking:
            peak_parking = current_parked

    if current_parked > capacity:
        status = "Alert: Parking Area is Full"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_parking)
    print("Parking Status:", status)

capacity = 50
parks = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

management_parking(capacity, parks)
```

```python
# Online Food Delivery Time Estimator
```

```python
def delivery_time(dist, traff, weather):
    
    base_time = dist * 5
    
    if traff == "Low":
        traff_delay = 0
    elif traff == "Medium":
        traff_delay = 5
    elif traff == "High":
        traff_delay = 10
    else:
        traff_delay = 0
    
    
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Rainy":
        weather_delay = 5
    elif weather == "Storm":
        weather_delay = 10
    else:
        weather_delay = 0
    
    
    eta = base_time + traff_delay + weather_delay
    print("Estimated Delivery Time:", eta, "minutes")

delivery_time(8, "High", "Rainy")
```

```python
#Movie Theatre Seat Occupancy Analyzer
```

```python
def seat_occupancy_analyse(total_seats, booked):
    
    booked_count = 0
    for seat in booked:
        booked_count += 1
    
    occupancy = (booked_count / total_seats) * 100
    
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Moderate Occupancy"
    else:
        status = "Low Occupancy"
    
    if occupancy >= 90:
        suggestion = "Suggestion: Open additional shows"
    else:
        suggestion = "Suggestion: No additional shows needed"
    
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print(suggestion)
total_seats = 200
booked = [1] * 150  
seat_occupancy_analyse(total_seats, booked)
```

```python
#Cloud Server Load Classification System
```

```python
def server_load(cpu_readings):
    total = 0
    for reading in cpu_readings:
        total += reading
    average = total / len(cpu_readings)
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)

cpu_readings = [45, 60, 70, 85, 90]
server_load(cpu_readings)
```

```python
#Smart Classroom Resource Usage Monitor
```

```python
def resource_usage(resource_usage):
    
    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
    else:
        print("Overused Resources: None")
    print("Energy Alert:", alert)
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
resource_usage(usage)
```

```python
#Online Event Registration Capacity Controller
```

```python
def manage_event(capacity, registrations):
    
    confirmed = 0
    waitlisted = 0
    
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


capacity = 100
registrations = 105

manage_event(capacity, registrations)
```
